```python
plt.figure(figsize=(12, 4))

# Plot the count of grades with respect to loan status
plt.subplot(1, 2, 1)
grade_order = sorted(df['grade'].unique())
sns.countplot(x='grade', hue='loan_status', data=df, order=grade_order, palette='coolwarm')
plt.title('Count of Grades with Respect to Loan Status')



# Adjust layout for better visualization
plt.tight_layout()

# Show the plot
plt.show()
```
